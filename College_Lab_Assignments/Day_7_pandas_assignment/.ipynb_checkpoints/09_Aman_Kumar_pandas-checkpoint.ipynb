```python
'''
14. Create a DataFrame my_movies containing movies that were released in the year 2000,
2004, 2008 and 2012.
'''
years = [2000, 2004, 2008, 2012]
my_movies = ratings[(ratings["type"] == "MOVIE") & (ratings["release_year"].isin(years))]
print(my_movies)
```

```python
'''
15. What is the mean, median and mode of the imdb_score column in the ratings
DataFrame?
'''
mean_imdb_score = ratings["imdb_score"].mean()
print(f"Mean IMDb score: {mean_imdb_score:.2f}")

median_imdb_score = ratings["imdb_score"].median()
print(f"Median IMDb score: {median_imdb_score:.2f}")

#mode_imdb_score = ratings["imdb_score"].mode()
#print(f"Mode IMDb score: {mode_imdb_score:.2f}")

from collections import Counter

mode_imdb_scores = Counter(ratings["imdb_score"]).most_common(1)

# Check if there are multiple modes
if len(mode_imdb_scores) > 1:
    print(f"Multiple modes: {mode_imdb_scores}")
else:
    print(f"Mode IMDb score: {mode_imdb_scores[0][0]} ({mode_imdb_scores[0][1]} occurrences)")
```

```python
'''
15. What is the mean, median and mode of the imdb_score column in the ratings
DataFrame?
'''
mean_imdb_score = ratings["imdb_score"].mean()
print(f"Mean IMDb score: {mean_imdb_score:.2f}")

median_imdb_score = ratings["imdb_score"].median()
print(f"Median IMDb score: {median_imdb_score:.2f}")

modes = ratings["imdb_score"].mode()

if len(modes) > 1:
  print(f"Multiple modes: {modes}")
else:
  print(f"Mode IMDb score: {modes.iloc[0]}")
```

```python
"""
16. What release years are represented in the dataset (must not include duplicates)?
"""
release_years = ratings["release_year"].unique()

print(f"Release years represented (without duplicates): {release_years}")
```

```python
"""
17. How often does each release year appear in the dataset?
"""
release_year_counts = ratings["release_year"].value_counts()

print(f"Release year counts:")
print(release_year_counts)
```

```python
"""
18. What are the most common runtime of the movies in the dataset?
"""
movies = ratings[ratings["type"] == "MOVIE"]
runtime_counts = movies["runtime"].value_counts()

#The idxmax() method on the Series returns the index of the element with the maximum value (highest count)
most_common_runtime = runtime_counts.idxmax()

print(f"Most common runtime for movies: {most_common_runtime} minutes")
```

```python
"""
19. Which are the top three mostly voted movies?
"""
top_3_rated_movies = (
    ratings.groupby("title")["imdb_score"].mean().nlargest(3)
    .reset_index()
    .sort_values(by="imdb_score", ascending=False)
)

print(f"Top 3 rated movies based on average IMDb score:")
print(top_3_rated_movies[["title", "imdb_score"]])
```

```python
"""
22. How many movies in the dataset are missing an age certification?
"""
missing_certifications = ratings["age_certification"].isna().sum()
print(f"Number of movies with missing age certifications: {missing_certifications}")
```

```python
"""
23. What are the most common age certifications in the dataset?
"""
certification_counts = ratings["age_certification"].value_counts()

# Check for missing values and handle them (if necessary)
if certification_counts.empty:
  print("No age certifications available in the dataset.")
else:
  most_common_cert = certification_counts.idxmax()
  print(f"Most common age certification: {most_common_cert} ({certification_counts[most_common_cert]} movies)")

# Show top N most frequent certifications
top_n_certs = certification_counts.nlargest(5)
print(f"Top {len(top_n_certs)} most common certifications:")
print(top_n_certs)
```
